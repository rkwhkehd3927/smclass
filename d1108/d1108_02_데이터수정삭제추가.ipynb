```python
import pandas as pd
data = {
   '이름' : ['강나래', '강태원', '강호림', '김수찬', '김재욱', '박동현', '박혜정', '승근열'],
   '학교' : ['구로고', '구로고', '구로고', '구로고', '구로고', '디지털고', '디지털고', '디지털고'],
   '키' : [197, 184, 168, 187, 188, 202, 188, 190],
   '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
   '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
   '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
   '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
   '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
   'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
df = pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df.index.name = '지원번호'
df.to_excel('score.xlsx')
df = pd.read_excel('score.xlsx')
```

데이터수정

```python
# replace: 데이터값을 변경, inplace=True 지정

df['학교'].replace({'구로고':'가산고'},inplace=True) # inplace=True : 반영
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)

# 특정 컬럼 값 변경
# 컬럼 = 변경값 입력
df['SW특기'] = df['SW특기'].str.lower() # 소문자
df['SW특기'] = df['SW특기'].str.upper() # 대문자
df['SW특기'] = df['SW특기'].str.capitalize() # 첫글자만 대문자
df['키'] = 200

# 컬럼명 = 문자 + 문자 일때만 가능
df['학교'] = df['학교']+'등학교'


# 특정셀 값 변경
df.loc['1번','2번'] = 20
df.loc['2번',['국어','영어']] = 10
```

```python
df.loc[['2번','3번'],['국어','영어']] = 50
df
```

```python
df['학교'] = df['학교']+'등학교'
df
```

```python
type(df['국어'])
df.info()
```

데이터추가

```python
## 데이터 추가
## 컬럼에 없는 컬럼 이름을 입력하면 컬럼이 추가됨
# df['합계'] = 0
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df['평균'] = df['합계']/5
```

```python
df
```

```python
# 결과 컬럼 추가
df['결과'] = 'Fail'


# 조건식을 사용해서 값을 변경
filt = df['합계']>=400
df.loc[filt,'결과'] = 'Pass'
df
```

```python
df['합산'] = 100

## 국어 점수가 90점 이상이면 df['합계']+100
filt = df['국어']>=90
df.loc[filt,'합산'] = df.loc[filt,'합산']+100
df
```

데이터 삭제

* columns
* index

drop - columns 삭제

```python
# 데이터 삭제 - drop: columns
df.drop(columns=['합산'],inplace=True)
# 컬럼 2개 삭제
df.drop(columns=['국어','영어'])

# 전체 컬럼의 슬라이싱을 가지고 특정 컬럼을 삭제할 수 있음
# df.columns[0]
df.columns[3:5]
df.drop(columns=df.columns[3:5])
```

```python
df.drop(columns=['합계','평균','결과'])
```

drop - index 삭제

```python
# drop 삭제 - index(row 값 삭제) : inplace = True 반영
df.drop(index='1번')
df.head(1)
```

```python
filt = df['수학']<80
filt
# 조건식을 이용한 삭제
df.drop(index=df[filt].index)
```

```python
# 김이 포함되어 있는 학생을 삭제
filt = df['이름'].str.contains('김')

df.drop(index=df[filt].index)
```
