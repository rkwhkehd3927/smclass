```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

다중막대그래프 - 옆으로 여러개 막대그래프

```python
df.shape
df
```

```python
# 배열 [0,1,2,3,4,5,6,7]
index = np.arange(df.shape[0])
```

```python
w = 0.25
plt.bar(index,df['국어'],width=w,label='국어')
plt.bar(index-w,df['영어'],width=w,label='영어')
plt.bar(index+w,df['수학'],width=w,label='수학')
plt.xticks(index)
plt.ylim(0,120)

# 값을 그래프에 출력
for i in range(8):
  plt.text(i,df['국어'][i]+0.2,df['국어'][i],ha='center')
  plt.text(i-w,df['영어'][i]+0.2,df['영어'][i],ha='center')
  plt.text(i+w,df['수학'][i]+0.2,df['수학'][i],ha='center')

plt.legend(ncol=3,loc='upper center',)
plt.show()
```
