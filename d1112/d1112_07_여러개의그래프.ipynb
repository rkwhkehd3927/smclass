```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
### x,y 축의 데이터
x=[1,2,3] # 리스트,튜플,DataFrame 등의 타입 - 리스트형태,Series
y=[2,4,8] # sm 매출액
y2 = [5,1,3] # LG 매출액
y3 = [1,2,5] # SK 매출액

plt.figure(figsize=(10,5))
plt.plot(x,y,label='SM매출액',marker='o')
plt.plot(x,y2,label='LG매출액',marker='x')
plt.plot(x,y3,label='SK매출액',marker='v',ls=':')

# 그래프 내 값 출력
for i in range(3):
  plt.text(x[i],y[i]+0.2,y[i],ha='center')
  plt.text(x[i],y2[i]+0.2,y2[i],ha='center')
  plt.text(x[i],y3[i]+0.2,y3[i],ha='center')


plt.title("기업별 매출액")
plt.legend(ncol=3) # 범례출력= 기본: 세로출력, ncol=3(한줄로 만들었음)
plt.show() 
```
