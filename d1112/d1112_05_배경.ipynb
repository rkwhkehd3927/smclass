```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x축: 지원번호, y축: 국어
# 선색상: red, marker 색상: yellow, 테두리:blue, 선타입 --

## x,y 축의 데이터
x = df.index
k = df['국어']
e = df['영어']


# 그래프 크기 조정, dpi= 그래프 해상도 설정(해상도 높을수록 이미지 크기 커짐), facecolor=배경색
plt.figure(figsize=(15,20),dpi=300,facecolor='yellow') # x축으로 5, y축으로는 20 크기 조절

## 선 그래프 생성
# linewidth= 선두께, color=선색상
# marker= 지점표시 - o,x,v, markersize(ms)=마커크기, markeredgecolor(mec)=마커테두리선색상, markerfacecolor(mfc)=마커색상
# alpha= 투명도설정, linestlye(ls)=라인스타일(:점선, --dash, -.)
# plt.plot(x,y, label='국어', color='red', marker='o', mfc='yellow ',mec='blue', ls='--')

plt.plot(x,k, label='국어', marker='o', color='red', mfc='yellow', mec='blue', ms=15, ls='--')
plt.plot(x,e, label='영어', marker='v', ms=15)


## 상단 타이틀
plt.title("국어성적",size=20) # size = 글자크기 변경

plt.legend(loc='upper right') # 좌표값으로도 위치 설정 가능

# 그래프 파일 저장
plt.savefig('graph2.png',dpi=300)
```
