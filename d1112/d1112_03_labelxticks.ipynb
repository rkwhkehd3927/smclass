```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x,y 축의 데이터
x=[-1,2,3] # 리스트,튜플,DataFrame 등의 타입 - 리스트형태,Series
y=[-2,-4,8]

# x = [1,2,3,4,5,6,7,8]
# x = df.index
# y = df['수학']

# 선 그래프 생성
plt.plot(x,y)

# 상단 타이틀
plt.title("라인그래프 Line",size=30) # size = 글자크기 변경
# X,Y축 설명, color = 글자색상(16코드,색상이름 전부 가능), loc = 글자위치
plt.xlabel('X축 날짜',color='#0303a1',loc='left') # 그래프 X축 설명
plt.ylabel('Y축 인원',color='red',loc='center') # 그래프 Y축 설명

# 그래프 눈금단위
plt.xticks([1,2,3])
plt.yticks([2,3,4,8,10])


plt.show()
```
