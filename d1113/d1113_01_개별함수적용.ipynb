```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.index
```

```python
plt.title('학생성적그래프')
plt.plot(df.index,df['국어'],label='국어')
plt.plot(df.index,df['영어'],label='영어')
plt.plot(df.index,df['수학'],label='수학')

for i in range(8):
  plt.text(df.index[i],df['국어'][i]+0.6,df['국어'][i],ha='center')
  plt.text(df.index[i],df['영어'][i]+0.6,df['영어'][i],ha='center')
  plt.text(df.index[i],df['수학'][i]+0.6,df['수학'][i],ha='center')

plt.legend()
```

```python
x=[1,2,3]
y=[2,4,8]
plt.plot(x,y)
plt.text(x[0],y[0],y[0])
```

```python
def func(x):
  if x-30>0: # 점수가 30보다 작으면 0 으로 출력
    return 0
  return x-30
  

# apply() = pandas 사용 문법
df['국어'].apply(func)
```

```python
# 누적막대그래프
plt.figure(figsize=(7,4))
plt.title('학생성적그래프')
plt.bar(df['이름'],df['국어'],width=0.6,label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],width=0.6,label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],width=0.6,label='수학')

# plt.legend(ncol=3)
n_list=df['국어'].apply(func) # 국어 점수 리스트

plt.legend()

for i in range(8): # text(x위치값,y위치값,값(숫자))
  plt.text(df['이름'][i],n_list[i],df['국어'][i],ha='center')
  # plt.text(df['이름'][i],df['국어'][i].apply(n_func),df['국어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]-20,df['영어'][i],ha='center') 
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i]-10,df['수학'][i],ha='center')
```
