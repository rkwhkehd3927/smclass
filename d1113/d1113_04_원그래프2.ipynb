```python
# 원그래프 pie()
values = [30,25,20,13,10,2] # 100퍼센트
# values = [1]*6 # 꼭 100퍼센트 비율로 맞추지 않아도 알아서 비율을 맞춰줌
explode = [0.1,0,0,0.1,0,0]
# wedgeprops=가운데공백(도넛형태), 'edgecolor':'테두리색상', 'linewidth':'테두리두께'
wedgeprops = {'width':0.8,'edgecolor':'white','linewidth':2}
labels = ['python','java','javascript','oracle','html','etc']
colors = ['mintcream','skyblue','plum','bisque','khaki','pink']
# pie(데이터값)(기본은 시계반대방향), labes=글자출력, explode= 원그래프 분리, counterclock=False 시계반대방향
# startangle=90(12시방향부터 시작)
# autopct=데이터 값 출력 '%.1f%%' : 소수점
# pctdistance=글자위치

### 데이터값 2% 이하의 값은 출력되지 않도록 하는 함수 선언
# pct = values 의 퍼센트 값 생성
# def custom_autopct(pct):
#   if pct >= 10:
#     # return ('%.1f%%' % pct)
#     # return ('{:.1f}%'.format(pct))
#     return (f'{pct:.1f}%') # 위에 두개랑 똑같음
#   else:
#     return ''
  
# 위에 함수 간단하게 만들기
def custom_autopct(pct):
  return f'{pct:.1f}%' if pct >= 10 else ''



plt.pie(values,labels=labels,autopct=custom_autopct,startangle=90, counterclock=False,
  colors=colors,wedgeprops=wedgeprops, pctdistance=0.7)
plt.show()
```
