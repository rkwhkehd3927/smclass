```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np
```

```python
### 남자 인구현황, 그래프 출력
df = pd.read_excel('202410연령별인구현황.xlsx',skiprows=3,nrows=1,usecols='B,E:Y',index_col='행정기관')
df.info()


### 여자 인구현황, 그래프 출력
df2 = pd.read_excel('202410연령별인구현황.xlsx',skiprows=3,nrows=1,usecols='B,AB:AV',index_col='행정기관')
# df.info()

# df.columns
# df.index # '전국  '

df2.info()
```

연령별 가로 막대그래프로 출력

```python
df = df.rename(index={'전국  ':'전국'})
df2 = df2.rename(index={'전국  ':'전국'})
```

```python
# 1. 데이터값, 천단위 표시 제거, int 타입 변경
df.iloc[0] = df.iloc[0].str.replace(",","").astype(int)
df2.iloc[0] = df2.iloc[0].str.replace(",","").astype(int)
```

```python
# 컬럼명 변경
df2.columns=df.columns.str.replace(".1","")
```

```python
# 컬럼명 변경 혹은 남자컬럼, 여자컬럼 복사
df2.columns = df.columns
```

```python

# 2. plt.bar(x,y) 출력
plt.barh(df.columns,-df.iloc[0]//1000,label='남자') # '-' 붙여주기
plt.barh(df2.columns,df2.iloc[0]//1000,label='여자')
plt.title('2024년 인구현황')
plt.xlabel('인구수')
plt.ylabel('연령대')

plt.legend()
```
