```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)

df
```

```python
# 컬럼 1개 - row 추가
df.loc[8] = ['평균',2024,df['관객 수'].mean(),df['평점'].mean()]
```

```python
plt.plot(df['영화'],[df['관객 수'][8]]*8,label='평균')
```

```python
# 컬럼 '평균' 을 추가하여 선그래프와 막대그래프 전부 출력
# 이름: 평균, 연도: 2024, 관객수: 평균관객수, 평점: 평균 평점
# df.loc[8] = ['평균',2024,df['관객 수'].mean(),df['평점'].mean()]
# 영화 관객수를 막대그래프로 출력
s_bar = plt.bar(df['영화'],df['관객 수'],width=0.6,label='관객 수')
plt.plot(df['영화'],[df['관객 수'][8]]*9,ls='--',color='black',label='평균')

plt.text(df['영화'][0],df['관객 수'][0],int(df['관객 수'][0]),ha='center') # 첫번째 거만 출력

plt.xticks(rotation=90)
plt.ylim(1000,1800)
s_bar[len(df.index)-1].set_color('red') # 평균 막대
plt.show()

  
```
