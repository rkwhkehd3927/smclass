```python
import pandas as pd
import numpy as np

df = pd.read_csv('chipotle.tsv',sep='\t')
```

```python
### 그룹화
df.groupby('학교').get_group('구로고')
df.groupby('학교').size() # 해당 학교 학생수 확인
df.groupby('학교').size()['구로고'] # 해당 학교 학생수 확인
df.groupby('학교')['키'].mean() # 학교별 '키'의 평균값 출력
```

```python
df.groupby('학교').count()
```

```python
df['item_price'].unique()
```

```python
# item_name 으로 그룹핑해서 개수를 나열한 것
df.groupby('item_name').value_counts()
```

그룹별 최대값,최소값,평균,합계

```python
# 학교 그룹별 - 키,국어점수 평균 출력
df.groupby('학교')[['키','국어']].mean()
df.groupby('학교')['키'].max()
```

```python
# 컬럼을 추가
df['학년'] = [3,3,2,1,3,2,2,1]
df
```

```python
df.groupby('학교').count()
df.groupby('학년').count()
df.groupby('학교').sum().sort_values('키')
```

```python
### 그룹화
# 컬럼을 기준으로 그룹핑 df.groupby(컬럼) - sum(): 합계, count(): 개수, size(): 크기
df.groupby[['학교','학년']].count()
# 컬럼지정필요: df.groupby(컬럼)[숫자타입컬럼] mean(),max(),min()
df.groupby('학교')['키'].min()
df.groupby('학교')[['키','국어']].min()

### 그룹별 키로 정렬,ascending=False : 역순 정렬
df.groupby('학년')['키'].mean().sort_values(ascending=False)
```

```python
# '학교'로 그룹핑, '학년'으로 학생수 출력
## 그룹핑을 했을 때 사용되는 개수
school = df.groupby('학교')
```

```python
# 학교로그룹핑을 하여, '구로고' 학생만 학년으로 학생수 출력
df.groupby('학교')['학년'].value_counts().loc['구로고']
# normalize=True: 퍼센ㅌ\
df.groupby('학교')['학년'].value_counts(normalize=True).loc['구로고']
```

```python
# '구로고' 학생의 '키'의 합계를 구하시오
df[df.groupby['학교'] == '구로고']['키'].sum().sort_values() # 망함
```
