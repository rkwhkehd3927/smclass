```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# 관객수 * 10000, 관객수로 역순정렬
df['관객 수'] = df['관객 수']*10000
df['관객 수'].sort_values(ascending=False)
df
```

```python
# 개봉 연도에 '년도'를 추가해서 출력
df['개봉 연도'] = df['개봉 연도'].astype(str)+'연도'
df
```

```python
# 관객수 천단위(,)를 표시하여 출력
df['관객 수'] = '$'+df['관객 수'].map('{:,d}'.format)
df
```

```python
# $ 없애고, 천단위 표시 없애고, int 타입으로 변경
df['관객 수'] = df['관객 수'].str[1:].str.replace(",","").astype(int)
df
```

```python
df.info()
```

```python
# 추천 점수 컬럼을 추가: 추천 점수 = (관객수 * 평점) // 100
df['추천 점수'] = (df['관객 수']* df['평점'])//100
df
```

연도 이걸로 바꿔야함 [2014, 2015, 2017, 2014, 2015, 2017, 2013, 2015]

```python
# 연도를 그룹핑하여 관객수 출력
df.groupby('개봉 연도')['관객 수'].sum().sort_values(ascending=False)
```
