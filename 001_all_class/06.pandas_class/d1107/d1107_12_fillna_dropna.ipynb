```python
import pandas as pd
import numpy as np
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

* fillna: NaN 데이터를 입력값으로 채워줌
* dropna: NaN 데이터 삭제
- axis: 축 index - row 방향으로 삭제, columns - column 방향으로 삭제
- how: 어떤 경우에 삭제할지 결정 any - 1개라도 NaN값이 있을때 삭제, all - 전체가 NaN 일때 삭제

* np.nan: numpy 라이브러리 NaN 입력

### fillna(): NaN 데이터 값 처리

```python
# 결측치(NaN데이터)
# NaN데이터 채우기: fillna()
df.fillna(' ',inplace=True) # NaN데이터를 빈 공백으로 채우기 # inplace=True 를 해야 저장됨
df.fillna('없음',inplace=True) # NaN데이터를 빈 공백으로 채우기 
df['SW특기'].fillna('없음',inplace=True) # 해당 컬럼에만 적용
df
```

```python
# 학교컬럼에 NaN 데이터를 입력: np.nan
df['학교'] = np.nan
df.fillna('없음') # inplace=True, 모든 NaN 값에 해당데이터를 입력
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
# df['학교'] = np.nan
```

### dropna() : NaN 데이터 삭제

```python
# NaN 데이터 삭제: dropna()
df.dropna() # NaN 데이터가 있는 row를 모두 삭제
```

```python
# axis 축: index 방향으로 삭제
# NaN 4,5번 row 데이터 삭제
df.dropna(axis='index', how='any') # any: 어떤 것이든 한개라도 있으면 삭제
```

```python
df['학교'] = np.nan
df
```

```python
# dropna(): NaN데이터 삭제
# axis - 축 (index: row삭제 ,columns: column 삭제) 
# how default: any - NaN 값이 1개라도 있으면 삭제, all - 전부 NaN일 때 삭제됨
df.dropna(axis='columns')
```
