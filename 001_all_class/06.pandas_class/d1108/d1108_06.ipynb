```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
# 구로고->가산고, 디지털고->단지고 로 변경
# SW특기를 소문자로 변경
# 4번 sw특기 -> django
# 5번 sw특기 -> python
# 합계 컬럼 추가
# 결과 컬럼 추가 후, 270점 이상은 pass, 아니면 fail 출력
```

```python
# 구로고->가산고, 디지털고->단지고 로 변경
df['학교'].replace({'구로고':'가산고','디지털고':'단지고'},inplace=True)
df
```

```python
# SW특기를 소문자로 변경
df.rename(columns=str.lower) # 소문자
df.columns
```

```python
# 4번 sw특기 -> django
# 5번 sw특기 -> python
df.loc[['4번','5번'],'SW특기'] = ['django','python']
df
```

```python
# 합계 컬럼 추가
df['합계']=0
df['합계']=df['국어']+df['수학']+df['영어']+df['과학']+df['사회']
df
```

```python
# '결과' 컬럼 추가 후, 270점 이상은 pass, 아니면 fail 출력
df['결과'] = 'Fail'
# filt = df['합계']>=270
# df.loc[filt,'결과'] = 'Pass'
# df.loc[~filt,'결과'] = 'Fail'
# df
df.loc[df['합계']> 400,'결과'] = 'Pass'
```

```python
### 문자열 = 문자열 + 문자열
df['학교'] = df['학교']+'등학교'
```
