데이터 조건 검색

```python
# 조건식 생성
filt = df['키']>=185
filt
# 조건식 대입
df[filt]
df[~filt] # ~: not
df.loc[filt]
```

&:  and 연산자, |: or 연산자

문자열 함수: startwith, contains, replace, lower, upper

```python
# startswith: 해당값으로 "시작"되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains: 해당값이 "포함"되어 있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt,['이름','키']]

# isin(): 포함되어 있는지 확인
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower(): 해당값을 모두 소문자로 변환.
langs = ['Python','Java']
filt = df['SW특기'].str.lower().isin(langs)
df[filt]

# 포함이 되어있으면 모두 검색
#  - NaN 데이터 처리하지 않으면 error
filt = df['SW특기'].str.contains('Java',na=False)
```

컬럼수정 방법

```python
filt = df['SW특기'].str.lower().isin(['python','java'])
df[filt]

# 소문자로 변경 후 입력값이 포함되어 있는 정보 검색

### j가 포함되어 있는 정보 검색
filt = df['SW특기'].str.lower().str.contains('j',na=False)
df[filt]

### 해달 컬럼에 값을 넣으면 컬럼의 값이 수정됨
df['SW특기'] = df['SW특기'].str.lower()
```

null 값 처리 : fillna - null 값 채우기, dropna - null 값 삭제하기

```python
import numpy as np


# fillna: 모든 컬럼의 NaN 데이터를 모두변경, inplace = True 해야 적용됨
df.fillna('데이터없음',inplace=True)
```

```python
df=pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# dropna
# axis 축 방향으로 삭제: index(row 방향), columns(columns 방향)
# how: any (1개라도 있으면 모두 row나 columns 삭제), all(모두 NaN 데이터여야 삭제)
# axis = 'index','columns' / how = 'any','all'
df.dropna() # index, 해당되는 index(row)를 삭제 / any가 기본(?)
# df['학교'] = np.nan # NaN 값 넣기

df.dropna(axis='index',how='all') # 모두 NaN row가 없어서 삭제가 안됨
df.dropna(axis='columns',how='all') # 학교 컬럼이 모두 NaN 이기에 삭제됨
df.dropna(axis='index',how='any') # 학교 컬럼이 모두 NaN 이기에 모든 row 삭제
df.dropna(axis='columns',how='any') # 학교 컬럼이 모두 NaN 이기에 삭제됨
```
