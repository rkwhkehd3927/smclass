```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# 함수선언
df['요일'] = 0
df
```

```python
# 특정 셀 부분 수정 가능
df.loc['1번','요일'] = 5
df.iloc[1,9] = 2
# 컬럼 전체 수정
df['요일'] = 0
df
```

```python
### 랜덤으로 0~6사이의 랜덤숫자를 입력
import random

for i in range(8):
  df.iloc[i,9] = random.randint(0,6)
# df.shape # (8, 10) 8행, 10열
df
```

```python
def func(x):
  if x == 0: return '일요일'
  elif x == 1: return '월요일'
  elif x == 2: return '화요일'
  elif x == 3: return '수요일'
  elif x == 4: return '목요일'
  elif x == 5: return '금요일'
  elif x == 6: return '토요일'
```

```python
df['요일'] = df['요일'].apply(func)
df
```

```python
## 결과 컬럼 생성
df['결과'] = '불합격'

## 국어점수가 60점 이상이면 '합격'으로 수정
filt = df['국어'] >= 60 
df.loc[filt,'결과'] = '합격'
df
```

```python
df = pd.read_csv('drinks.csv')
df
```

```python
df.shape
df.info()
```

```python
# 모든 술의 총량을 출력
# total_servings 컬럼 구하기: beer+spirit+wine
# 전체 알코올의 도수 확인
df['total_servings'] = df['beer_servings']+df['spirit_servings']+df['wine_servings']
df
```

```python
# 대륙별 total_servings 구하기

# df['continent'].unique() # 'AS', 'EU', 'AF', nan, 'SA', 'OC' -> 대륙 6개
df.groupby('continent')['total_servings'].sum().sort_values() # nan 은 계산 안됨
```

```python
# null 에 들어간 값들 찾기
filt = df['continent'].isnull() 
df.loc[filt,'total_servings']
df.loc[filt,'total_servings'].sum()
```

```python
# continent: 대륙 NaN 값을 OT로 변경
df['continent'].fillna('OT',inplace=True)
```

```python
df.groupby('continent')['total_servings'].sum().sort_values() # 위에서 안됐던거(NaN) 다시 계산
```

```python
df.columns
```

```python
# 술 소비량 대비 알콜 비율이 어느정도 되는지 출력
# df.groupby('total_servings')[]
# 알코올 비율을 계산해서 컬럼(alcohol_rate) 추가
df['alcohol_rate'] = df['total_litres_of_pure_alcohol'] / df['total_servings']
# total_litres_of_pure_alcohol(알코올비율)을 total_servings으로 나눈 것
# 술소비량 대비 알콜 비율이 높은 나라별 20개를 출력 (높은순) - head(20)

# 대륙별 알콜비율 계산해서 출력 (평균)
df.groupby('continent')['alcohol_rate'].sum().sort_values(ascending=False)
```

```python
df['alcohol_rate'].fillna(0,inplace=True)
df
```
