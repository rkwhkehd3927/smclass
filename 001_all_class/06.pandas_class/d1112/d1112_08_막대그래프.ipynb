```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

라벨 변경, 라벨의 기울기 설정, width, alpha 투명도 조정


```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors = ['r','g','b']


plt.figure(figsize=(5,3)) # 그래프 크기 조절

# plt.plot(label,values) # 선그래프
# plt.bar(label,values,color='r') # 바그래프, 1개의 색상으로 처리
plt.bar(label,values,color=colors, width=0.2, label='학생 키') # 바색상을 여러 개로 처리

plt.yticks([0,50,100,150,200,250])
# plt.xticks([0,1,2],rotation=45)
plt.xticks(label,['1번','2번','3번'],rotation=45) # x라벨 변경

plt.title("막대 그래프") # 상단 제목
plt.legend(loc='upper center')
# plt.legend(loc=(0.5,0.5))
plt.show() 
```

막대그래프의 무늬설정 : set_hatch()

```python
colors= ['red','green','blue']
bar = plt.bar(label,values,width=0.5,color=colors)
# 막대그래프의 무늬설정
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
```

* 범위설정 xticks, yticks = 0 부터 표시됨
* 특정구간부터 범위 설정 xlim, ylim

```python
colors= ['red','green','blue']
plt.barh(label,values)
plt.yticks([0,100,150,200]) # 0부터 범위 지정
plt.ylim(175,190) # 특정 구간 범위 지정
plt.show()
```

```python
label = ['홍길동','유관순','강감찬']
values = [180,187,184]
colors= ['red','green','blue']
plt.barh(label,values,height=0.5,color=colors)
plt.show()
```
