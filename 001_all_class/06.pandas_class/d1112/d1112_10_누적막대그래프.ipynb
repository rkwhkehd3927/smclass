```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

누적막대그래프
* bottom = 첫번째 값을 입력

```python
df
```

```python
plt.figure(figsize=(7,4))
plt.bar(df['이름'],df['국어'],label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],label='영어')
plt.xticks(rotation=45)

# 그래프 값
for i in range(8):
  plt.text(df['이름'][i],df['국어'][i],df['국어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i],df['영어'][i],ha='center')
  plt.text(df['이름'][i],df['국어'][i]+df['영어'][i]+df['수학'][i],df['수학'][i],ha='center')


plt.legend(ncol=3,loc='upper center') # 범례
plt.ylim(50,350) # y축 범위
plt.show()
```
