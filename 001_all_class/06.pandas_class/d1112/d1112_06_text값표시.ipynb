그래프에 값 표시

```python
### x,y 축의 데이터
x=[1,2,3] # 리스트,튜플,DataFrame 등의 타입 - 리스트형태,Series
y=[2,4,8]

plt.figure(figsize=(7,5))
plt.plot(x,y,label='범례데이터',marker='o')
## text (x좌표 위치, y좌표 위치, 데이터값), ha=글자를 가운데정렬
# plt.text(x[0],y[0],y[0],ha='center')
# plt.text(x[1],y[1],y[1],ha='center')

# 그래프에 값 표시
for idx,txt in enumerate(y):
  plt.text(x[idx]+0.05,y[idx]-0.4,y[idx],ha='center',color='red')


plt.title('라인그래프')
plt.legend()
plt.show()
```
