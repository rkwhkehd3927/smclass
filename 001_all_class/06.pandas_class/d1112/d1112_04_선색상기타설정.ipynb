```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
## x,y 축의 데이터
x=[-1,2,3] # 리스트,튜플,DataFrame 등의 타입 - 리스트형태,Series
y=[-2,-4,8]
yy=[3,5,9]

## 선 그래프 생성
# linewidth= 선두께, color=선색상
# marker= 지점표시 - o,x,v, markersize(ms)=마커크기, markeredgecolor(mec)=마커테두리선색상, markerfacecolor(mfc)=마커색상
# alpha= 투명도설정, linestlye(ls)=라인스타일(:점선, --dash, -.)

plt.plot(x,y, label='인원변동', color='pink', linewidth=5, marker='o', ms=10, mec='red')
# plt.plot(x,yy, label='비용변동',marker='v', ms=10, mfc='yellow', alpha=0.3, ls='--')

# b=선색상, v=마커표시, =선모양  == 'bv:' 단축표시가능
plt.plot(x,yy, 'bv:', label='비용변동')

## 상단 타이틀
plt.title("라인그래프 Line",size=20) # size = 글자크기 변경

## 범례: 그래프 선 설명문구, loc= upper, center, lower, left, right, 
# plt.legend(loc='center')
plt.legend(loc=(0.2,0.5)) # 좌표값으로도 위치 설정 가능

plt.show()
```
