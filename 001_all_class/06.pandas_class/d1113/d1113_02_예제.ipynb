```python
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # window용 # 안넣으면 한글 깨짐
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 특수기호 (-) 출력
# matplotlib.rcParams['font.family'] = 'AppleGothic'   # Mac용
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
df.index
```

```python
def cm_func(x):
  return str(x)+'cm'


# 배열타입의 모든것을 함수적용 = apply
df['키'] = df['키'].apply(cm_func)
df['키']
```

```python
# 50점 아래에 있는 값을 50으로 리턴하여 출려
def kor_func(x):
  if x<50:
    return 50
  return x
  

# int, float,str 단일 값은 함수명(데이터)로 호출
for i in range(8):
  print(kor_func(df['국어'][i]))
```

```python
def h_func(x):
  return x/2

plt.bar(df['이름'],df['국어'],width=0.6,label='국어')
plt.bar(df['이름'],df['영어'],bottom=df['국어'],width=0.6,label='영어')
plt.bar(df['이름'],df['수학'],bottom=df['국어']+df['영어'],width=0.6,label='수학')

# 단일타입의 데이터 - int,float,str
for i in range(8):
  plt.text(df['이름'][i],h_func(df['국어'][i]),df['국어'][i],ha='center') # 값의 y축 위치를 바의 중간에 배치
  plt.text(df['이름'][i],h_func(df['국어'][i]+df['국어'][i]+df['영어'][i]),df['영어'][i],ha='center') # 값의 y축 위치를 바의 중간에 배치
  plt.text(df['이름'][i],h_func(df['국어'][i]+df['영어'][i]+df['국어'][i]+df['영어'][i]+df['수학'][i]),df['수학'][i],ha='center') # 값의 y축 위치를 바의 중간에 배치


# 누적막대그래프를 사용해서 값을 모두 중앙에 출력

```
