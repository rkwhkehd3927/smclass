```python
# 원그래프 pie()
values = [30,25,20,13,10,2] # 100퍼센트
# values = [1]*6 # 꼭 100퍼센트 비율로 맞추지 않아도 알아서 비율을 맞춰줌
explode = [0.1,0,0,0.1,0,0]
labels = ['python','java','javascript','oracle','html','etc']
colors = ['mintcream','skyblue','plum','bisque','khaki','ivory']
plt.title('프로그램 사용률')
# pie(데이터값)(기본은 시계반대방향), labes=글자출력, explode= 원그래프 분리, counterclock=False 시계반대방향
# startangle=90(12시방향부터 시작)
# autopct=데이터 값 출력 '%.1f%%' : 소수점
plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%',counterclock=False,startangle=90,colors=colors)
plt.legend(loc=(1.2,0.3),title='언어별 선호도')
plt.show()
```
